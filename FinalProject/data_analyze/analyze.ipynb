```python
import pandas as pd
import matplotlib.pyplot as plt


code_path = '../data/sample_codes_stat@0.csv'
commit_path = '../data/sample_commits@0.csv'
```

## 1. Hypothesis 1:
####     The more complex the code is, the harder it is to discover its bug(s).

```python
code_stat = pd.read_csv(code_path,index_col=['bug'])
commit = pd.read_csv(commit_path,index_col=['bug'])
df = pd.merge(commit, code_stat, on='bug')
```

```python
fig,ax = plt.subplots(1,5,figsize=(30,5),)
x = 'find_bug_time'
j = 0
for y in ['nloc','CCN', 'function_numbers', 'token_count', 'parameter_count']:
    df.plot(kind="scatter",x=x,y=y,ax=ax[j],title='plot between {x} and {y}'.format(x=x,y=y))
    j += 1
```

## 2. Hypothesis 2:
####     The more complex the code is, the harder it is to fix its bug(s).

```python
fig,ax = plt.subplots(2,5,figsize=(30,10),)
i = 0
for x in ['fix_distance',  'fix_bug_time']:
    j = 0
    for y in ['nloc','CCN', 'function_numbers', 'token_count', 'parameter_count']:
        df.plot(kind="scatter",x=x,y=y,ax=ax[i][j],title='plot between {x} and {y}'.format(x=x,y=y))
        j += 1
    i += 1   
```
